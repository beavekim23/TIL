```python
query = """
SELECT *
    FROM Customers
;
"""
df = pd.read_sql(query, db)
```

```python
fran = df['Country'] == "France"
germ = df['Country'] == "Germany"
df[fran | germ]
```

```python
query = """
SELECT *
    FROM Customers
    WHERE 
        Country IN ('France', 'Germany')
"""

pd.read_sql(query, db)
```
