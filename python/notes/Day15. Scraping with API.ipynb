```python
for item in results:
    news_link = item.get_attribute('href')
    subject = item.find_element_by_class_name('subject').text
    date = item.find_element_by_class_name('date').text
    abs_text = item.find_element_by_class_name('txt').text
    
    print(news_link, subject, date, abs_text, sep="\n")
    print("\n")
```

```python
def get_news_list():
    result_list = []
    for item in results:
        news_link = item.get_attribute('href')
        subject = item.find_element_by_class_name('subject').text
        date = item.find_element_by_class_name('date').text
        abs_text = item.find_element_by_class_name('txt').text

        print(news_link, subject, date, abs_text, sep="\n")
        print("\n")
        result_list.append((
            news_link, subject, date, abs_text
        ))
    return result_list
```
