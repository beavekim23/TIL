## Review

```python
for i in range(1, 100+1):
    if i % 15 == 0:
        print("fizzbuzz")
    elif i % 3 == 0:
        print("fizz")
    elif i % 5 == 0:
        print("buzz")
    else:
        print(i)
```

```python
#fizzbuzz 다른 방법(elif 안쓰기)
#"string"*True는 string, "string"*False는 ""임을 이용

for i in range(1, 100+1):
    if i%3==0 or i%5==0:
        result = "fizz"*(i%3==0) + "buzz"*(i%5==0)
        print(result)
    else:
        result = i
        print(result)
```

## Monty Hall Problem

```python
import random


stay = 0
switch = 0
trial = 10

for _ in range(trial):
    doors = ["car", "goat", "goat"]
    random.shuffle(doors)
    #print(doors)
    choice = random.randrange(3)
    user = doors.pop(choice)
    #print(user)
    doors.pop(doors.index("goat"))
    #goat 값이 있는 index찾아서 pop
    
    if user == "car":
        stay += 1
    else:
        switch += 1
    
print("stay: {} {}, switch: {} {}".format(stay, stay/trial, switch, switch/trial))
    
```
