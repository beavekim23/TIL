## Assignment
Select the whole data from Orders where<br/>
 1) Quantity >= 40 in Orderdetails <br/>
 2) CustomerName contains "Ernst" or "Stop" in Customers<br/>

```python
#Inspect Orders table

query = """
SELECT * 
FROM Orders;
"""

pd.read_sql(query, db).head()
```

```python
# Inspect OrderDetails table under the first condition

query = """
SELECT *
FROM OrderDetails
WHERE
    Quantity >= 40
;
"""

pd.read_sql(query, db).head()
```

```python
# Apply the first condition

query = """
SELECT *
FROM Orders
WHERE 
    OrderID IN(
        SELECT OrderID
        FROM OrderDetails
        WHERE
        Quantity >= 40
    )
"""

pd.read_sql(query, db).head()
```

```python
# Inspect Customers Table

query = """
SELECT *
FROM
    Customers
;
"""

pd.read_sql(query, db).head()
```

```python
# Inspect Customers table under the second condition

query = """
SELECT CustomerID
FROM
    Customers
WHERE
    CustomerName LIKE '%Ernst%'
    OR
    CustomerName LIKE '%Stop%'
;
"""

pd.read_sql(query, db).head()
```

```python
# finally

query = """
SELECT *
FROM Orders
WHERE 
    OrderID IN(
        SELECT OrderID
        FROM OrderDetails
        WHERE
            Quantity >= 40
    )
    AND
    CustomerID IN(
        SELECT CustomerID
            FROM
                Customers
            WHERE
                CustomerName LIKE '%Ernst%'
                OR
                CustomerName LIKE '%Stop%'
        )
"""

pd.read_sql(query, db)
```

## Join with SQL

```python
#Do not do this

query = """
SELECT *
FROM
    Customers, Orders
;
"""

pd.read_sql(query, db).head()
```

```python
#Join with sql 1 : without using JOIN

query = """
SELECT C.CustomerID, C.CustomerName, O.CustomerID, O.OrderID
FROM
    Customers C, Orders O
WHERE
    C.CustomerID = O.CustomerID
;
"""

pd.read_sql(query, db).head()
```

```python
#Join with sql 2 : using JOIN, ON

query = """
SELECT C.CustomerID, C.CustomerName, O.CustomerID, O.OrderID
FROM Customers C
    JOIN
        Orders O
    ON
        C.CustomerID = O.CustomerID
;
"""
# len(pd.read_sql(query, db))
pd.read_sql(query, db).head()
```

## GROUP BY with Pandas : use .groupby(), .get_group()

```python
query = """
SELECT *
FROM
    Orders
;
"""

orders_df = pd.read_sql(query, db)
orders_df.head()
```

```python
date_group = orders_df.groupby("OrderDate")
```

```python
date_group.get_group("1996-07-08")
```

```python
orders_df["OrderDate"].unique()
```

```python
orders_df["OrderDate"].nunique()
```

```python
#count the number of orders using group by with pandas

daily_order_count = pd.DataFrame([
    {
        "OrderDate": OrderDate,
        "Count" : len(date_group.get_group(OrderDate))
    } for OrderDate in orders_df["OrderDate"].unique()
    
])

daily_order_count.head()
```

## GROUP BY with SQL

```python
query = """
SELECT count(*) "Count", OrderDate
FROM
    Orders
    GROUP BY
        OrderDate
;
"""

pd.read_sql(query, db).head()
```
